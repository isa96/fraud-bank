# Fraud Bank Detection
In this project, the dataset that I'm using is from BCA Indonesia. Using Random Forest, I will classificate fraud detection.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df=pd.read_csv('fraud_train.csv')
print(df.info())
df.head()
```

```python
def nullity(data):
    df=pd.DataFrame(columns=['dataFeatures','dataType','null','nullPct','unique','uniqueSample'])
    df['dataFeatures']=data.columns
    df['dataType']=[i for i in data.dtypes]
    df['null']=df['dataFeatures'].apply(lambda x: sum(data[x].isna()))
    df['nullPct']=df['null'].apply(lambda x: round((x/len(data))*100,2))
    df['unique']=df['dataFeatures'].apply(lambda x: len(data[~(data[x].isna())][x].unique()))
    df['uniqueSample']=df['dataFeatures'].apply(lambda x: data[x].drop_duplicates().sample(1).values)
    return df
nullity(df)
```

### Filling NaN values

```python
idx_error=df[df['minimum_nilai_transaksi'].isna()].index

for i in idx_error:
    df.loc[i, 'rata_rata_jumlah_transaksi']=1.00
    for j in ['rata_rata_nilai_transaksi', 'maksimum_nilai_transaksi', 'minimum_nilai_transaksi']:
        df.loc[i, j]=df.loc[i,'nilai_transaksi']
```

### Fraud cases from each city

```python
matrix = []
for i in df['nama_kota'].unique():
    if len(df[df['nama_kota']==i]) >= 20:
        matrix.append([i, 
                       len(df[df['nama_kota']==i]), 
                       len(df[(df['nama_kota']==i) & (df['flag_transaksi_fraud'] == 1)]), 
                       len(df[(df['nama_kota']==i) & (df['flag_transaksi_fraud'] == 0)]), 
                       (len(df[(df['nama_kota']==i) & (df['flag_transaksi_fraud'] == 1)]))/ len(df[df['nama_kota'] == i])])

pd.DataFrame(matrix, columns = ['Kota', 'Banyak Transaksi', 'Banyak Fraud', 'Banyak Normal', 'Persentasi Fraud']).sort_values(by = 'Persentasi Fraud', ascending = False).head(10)    
```

```python
for i in df.select_dtypes(exclude='float').columns:
    df[i]=df[i].astype('object')
```

```python
df['selisih_transaksi_rata_rata']=df.apply(lambda x: x['nilai_transaksi']-x['rata_rata_nilai_transaksi'], axis=1)
```

### Feature Importance based on chi2

```python
from scipy.stats import chi2_contingency

chi2_check=[]
for i in df.select_dtypes('object').columns:
    if chi2_contingency(pd.crosstab(df['flag_transaksi_fraud'], df[i]))[1]<0.05:
        chi2_check.append('Reject Ho')
    else:
        chi2_check.append('Fail to reject Ho')
        
res=pd.DataFrame(data={'Column':df.select_dtypes('object').columns, 'Hypothesis':chi2_check})
res
```

```python
df.drop(columns=['X','id_tanggal_transaksi_awal','id_merchant','nama_merchant',
                'id_negara','id_channel','flag_transaksi_finansial','status_transaksi',
                'bank_pemilik_kartu','tipe_mesin'], inplace=True)
```

## Random Forest Classifier with RandomizedSearchCV

```python
from sklearn.ensemble import RandomForestClassifier

data=df.drop(['flag_transaksi_fraud'], axis=1)
target=df['flag_transaksi_fraud'].astype('int')
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(data, target, test_size=0.3)
```

### Randomized Search CV

```python
from sklearn.model_selection import RandomizedSearchCV

rfc_=RandomForestClassifier()

#number of tree
n_estimators=[int(x) for x in np.linspace(start=200, stop=2000, num=10)]

#number of features to consider every split
max_features=['auto','sqrt']

#maximum number of levels in tree
max_depth=[int(x) for x in np.linspace(10,110,num=11)]
max_depth.append(None)

#Minimum number of samples required to split a node
min_samples_split=[int(x) for x in np.linspace(start=100, stop=1000, num=100)]

#Minimum number of samples required at each leaf node
min_samples_leaf=[int(x) for x in np.linspace(1,11,num=10)]

#method of selecting samples for training each tree
bootstrap=[True, False]

#create the random grid
random_grid={'n_estimators':n_estimators,
            'max_features':max_features,
            'max_depth':max_depth,
            'min_samples_split':min_samples_split,
            'min_samples_leaf':min_samples_leaf,
            'bootstrap':bootstrap}
```

```python
random=RandomizedSearchCV(estimator=rfc_,
                         param_distributions=random_grid,
                         n_iter=20, cv=3, n_jobs=-1, scoring='accuracy')
```

```python
random.fit(x_train, y_train)
```

```python
random.best_params_
```

```python
rfc=RandomForestClassifier(n_estimators=1000, min_samples_split=118,
                           min_samples_leaf=6, max_features='sqrt',
                           max_depth=30, bootstrap=True)
rfc.fit(x_train, y_train)
```

```python
from sklearn.metrics import classification_report

fraud_pred=rfc.predict(x_test)
print(classification_report(y_test, fraud_pred))
```

## Applying Threshold

```python
pred_prob=rfc.predict_proba(x_test)

plt.figure(figsize=(8,6))
sns.distplot(pred_prob[:,0], label='Not Fraud')

plt.xlabel('Probability')
```

```python
final_pred=[]
for i in pred_prob[:,0]:
    if i>0.95:
        final_pred.append(0)
    else:
        final_pred.append(1)
        
final_pred=np.array(final_pred)
print(classification_report(y_test, final_pred))
```
